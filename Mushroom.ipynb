```python
# Prediction
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
x=data.iloc[:,data.columns!='class']
y=data.iloc[:,0]
print(y)
```

```python
lab_enc=preprocessing.LabelEncoder()
y_fit=lab_enc.fit_transform(y)
y_fit
```

```python
x_enc=pd.get_dummies(x)
x_enc.head(15)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_enc,y_fit,test_size=0.2,random_state=0,)
```

```python
print("x Train Shape: ",x_train.shape)
print("y Train Shape: ",y_train.shape)
print("\nx Test Shape: ",x_test.shape)
print("y Test Shape: ",y_test.shape)
```

```python
model_k3=KNeighborsClassifier(n_neighbors=3)
model_k5=KNeighborsClassifier(n_neighbors=3)
model_k3.fit(x_train,y_train)
model_k5.fit(x_train,y_train)
y_pred_k3=model_k3.predict(x_test)
y_pred_k5=model_k5.predict(x_test)
```

```python
df2 = pd.DataFrame({'y test':y_test, 'y prediction k=3':y_pred_k3,'y prediction k=5':y_pred_k5})
df2
```

```python
print("Accuracy Score for k=3",(accuracy_score(y_pred_k3,y_test)*100))
print("Accuracy Score for k=5",(accuracy_score(y_pred_k5,y_test)*100))
```
